#Load the accompanying data into a df.

```python
# Mounting google colab, this will prompt first time each session.
drive.mount('/content/drive',force_remount=True)
dataset_file = "/content/drive/My Drive/Colab Notebooks/train.csv"
df=pd.read_csv(dataset_file,sep=',')
df.name="Original"
print(df.head(10))
print(df.shape)
```

```python
patients = df['patient_id'].nunique()
images = df['image_id'].nunique()
implants = sum(df['implant'])
cancer = sum(df['cancer'])
cancer_and_implants = sum((df['cancer']==1) & (df['implant']==1))
cc_images = sum((df['view']=="CC"))
mlo_images = sum((df['view']=="MLO"))

cc_and_cancer = sum((df['view']=="CC") & (df['cancer']==1))
mlo_and_cancer = sum((df['view']=="MLO") & (df['cancer']==1))

cc_no_implants = sum((df['view']=="CC") & (df['implant']==0))
cc_and_cancer_no_implants = sum((df['view']=="CC") & (df['cancer']==1) & (df['implant']==0))
cc_and_cancer_and_implants= sum((df['view']=="CC") & (df['cancer']==1) & (df['implant']==1))

print(f"Total Images: '{images}' Total Patients: '{patients}'")
print(f"Images with Cancer: '{cancer}'")

print(f"Images with Implants: '{implants}'")
print(f"Images with Cancer: '{cancer}'")
print(f"Images with Cancer & Implant: '{cancer_and_implants}'")

print(f"CC Images: '{cc_images}'")
print(f"CC Images with Cancer: '{cc_and_cancer}'")
print(f"CC Images with Cancer & Implant: '{cc_and_cancer_and_implants}'")

print(f"MLO Images: '{mlo_images}'")
print(f"MLO Images with Cancer: '{mlo_and_cancer}'")

print(f"CC Images, no implants: '{cc_no_implants}'")
print(f"CC Images, no implants,  with Cancer: '{cc_and_cancer_no_implants}'")

print(f"Percentage of CC Images: '{round(cc_images/images*100,2)}%'")
print(f"Percentage of MLO Images: '{round(mlo_images/images*100,2)}%'")

print(f"Percentage of Images with Implants: '{round(implants/images*100,2)}%'")
print(f"Percentage of Images with Cancer: '{round(cancer/images*100,2)}%'")
print(f"Percentage of Images with Cancer among those with implants: '{round(cancer_and_implants/implants*100,2)}%'")
print(f"Percentage of CC Images with Cancer among CC Images: '{round(cc_and_cancer/cc_images*100,2)}%'")
print(f"Percentage of MLO Images with Cancer among MLO Images: '{round(mlo_and_cancer/mlo_images*100,2)}%'")

print(f"Percentage of CC Images(no implants) with Cancer among CC Images(no implants): '{round(cc_and_cancer_no_implants/cc_no_implants*100,2)}%'")
```
